```python
import pandas as pd
import random
from faker import Faker
from datetime import datetime, timedelta

# Initialize Faker and seed
fake = Faker()
Faker.seed(303)
random.seed(303)

# Load necessary tables
stores_df = pd.read_csv("Stores.csv")
employees_df = pd.read_csv("Employees.csv")
customers_df = pd.read_csv("Customers.csv")
products_df = pd.read_csv("Products.csv")
promotions_df = pd.read_csv("Promotions.csv")


import pandas as pd
import random
from faker import Faker
from datetime import datetime

fake = Faker()

# Assuming these DataFrames already exist from earlier in your notebook:
# stores_df, employees_df, customers_df, promotions_df, products_df



# ------------------ SALES ------------------
import pandas as pd
import random
from faker import Faker
from datetime import datetime

fake = Faker()

# Assuming these DataFrames are preloaded:
# stores_df, employees_df, customers_df, promotions_df, products_df, vendors_df (for VendorID)

# ------------------ SALES ------------------
sales = []
sale_details = []
payments = []

sale_id_start = 7001
saledetail_id = 8001
payment_id = 9001

payment_methods = ["Cash", "Credit", "Debit", "EBT", "Mobile"]
payment_types = ["Customer", "Vendor", "Refund"]

for sale_index in range(550):
    sale_id = sale_id_start + sale_index
    store = stores_df.sample(1).iloc[0]
    employee = employees_df[employees_df["StoreID"] == store["StoreID"]].sample(1).iloc[0]
    customer = customers_df.sample(1).iloc[0]
    promotion = promotions_df.sample(1).iloc[0] if random.random() < 0.3 else None
    sale_date = fake.date_time_between(start_date='-30d', end_date='now')

    num_items = random.randint(1, 4)
    sale_total = 0.0
    used_product_ids = set()

    for _ in range(num_items):
        while True:
            product = products_df.sample(1).iloc[0]
            if product["ProductID"] not in used_product_ids:
                used_product_ids.add(product["ProductID"])
                break

        quantity = random.randint(1, 5)
        unit_price = product["UnitPrice"]
        discount = 0.0

        if promotion is not None and int(promotion["ProductID"]) == int(product["ProductID"]):
            discount = (promotion["DiscountPercentage"] / 100) * unit_price

        line_total = (unit_price - discount) * quantity
        sale_total += line_total

        sale_details.append({
            "SaleDetailID": saledetail_id,
            "SaleID": sale_id,
            "ProductID": product["ProductID"],
            "Quantity": quantity,
            "UnitPrice": unit_price,
            "DiscountAmount": round(discount * quantity, 2)
        })
        saledetail_id += 1

    pm = random.choice(payment_methods)
    pt = random.choices(
        population=payment_types,
        weights=[0.8, 0.15, 0.05],  # Mostly Customer, some Vendor, few Refunds
        k=1
    )[0]

    sales.append({
        "SaleID": sale_id,
        "StoreID": store["StoreID"],
        "EmployeeID": employee["EmployeeID"],
        "CustomerID": customer["CustomerID"],
        "PromotionID": int(promotion["PromotionID"]) if promotion is not None else None,
        "SaleDate": sale_date,
        "TotalAmount": round(sale_total, 2),
        "PaymentMethod": pm,
        "ChangeGiven": 0.0 if pm != "Cash" else round(random.uniform(0, 5), 2)
    })

    # Assign VendorID only for Vendor payment type
    vendor_id = vendors_df.sample(1).iloc[0]["VendorID"] if pt == "Vendor" else pd.NA

    payments.append({
        "PaymentID": payment_id,
        "PaymentType": pt,
        "VendorID": vendor_id,
        "SaleID": sale_id,
        "DeliveryID": pd.NA,
        "PaymentDate": sale_date.strftime("%Y-%m-%d"),
        "PaymentMethod": pm,
        "Amount": round(sale_total, 2),
        "ReferenceNumber": fake.uuid4()[:8]
    })
    payment_id += 1

# Convert to DataFrames
sales_df = pd.DataFrame(sales)
sale_details_df = pd.DataFrame(sale_details)
payments_df = pd.DataFrame(payments)

# 🔧 Type safety
sales_df["PromotionID"] = sales_df["PromotionID"].astype("Int64")
sale_details_df["ProductID"] = sale_details_df["ProductID"].astype("int")
payments_df["SaleID"] = payments_df["SaleID"].astype("int")

# Save to CSV
sales_df.to_csv("Sales.csv", index=False)
sale_details_df.to_csv("SaleDetails.csv", index=False)
payments_df.to_csv("Payments.csv", index=False)
```

```python
create_tables_sql = r"""
DROP TABLE IF EXISTS Payments CASCADE;
DROP TABLE IF EXISTS Promotions CASCADE;
DROP TABLE IF EXISTS EmployeeShift CASCADE;
DROP TABLE IF EXISTS Inventory CASCADE;
DROP TABLE IF EXISTS SaleDetails CASCADE;
DROP TABLE IF EXISTS Sales CASCADE;
DROP TABLE IF EXISTS DeliveryDetails CASCADE;
DROP TABLE IF EXISTS Deliveries CASCADE;
DROP TABLE IF EXISTS Customers CASCADE;
DROP TABLE IF EXISTS StoreManagers CASCADE;
DROP TABLE IF EXISTS Employees CASCADE;
DROP TABLE IF EXISTS Stores CASCADE;
DROP TABLE IF EXISTS Products CASCADE;
DROP TABLE IF EXISTS Vendors CASCADE;
DROP TABLE IF EXISTS ProductCategories CASCADE;

-- Product Categories
CREATE TABLE ProductCategories (
    CategoryID SERIAL PRIMARY KEY,
    CategoryName VARCHAR(50) NOT NULL UNIQUE
);

-- Vendors
CREATE TABLE Vendors (
    VendorID SERIAL PRIMARY KEY,
    VendorName VARCHAR(100) NOT NULL,
    ContactName VARCHAR(100) NOT NULL,
    PhoneNumber VARCHAR(15) NOT NULL,
    Email VARCHAR(100) NOT NULL CHECK (Email ~* '^[A-Za-z0-9._%-]+@[A-Za-z0-9.-]+[.][A-Za-z]+$'),
    Address TEXT NOT NULL
);

-- Products
CREATE TABLE Products (
    ProductID SERIAL PRIMARY KEY,
    ProductName VARCHAR(100) NOT NULL,
    Aisle VARCHAR(10),
    Shelf VARCHAR(10),
    CategoryID INT NOT NULL REFERENCES ProductCategories(CategoryID),
    UnitPrice DECIMAL(10,2) NOT NULL CHECK (UnitPrice > 0),
    VendorID INT NOT NULL REFERENCES Vendors(VendorID),
    SKU VARCHAR(50) UNIQUE,
    IsActive BOOLEAN DEFAULT TRUE,
    LastRestockedDate TIMESTAMPTZ,
    CONSTRAINT valid_aisle_format CHECK (Aisle ~* '^[A-Z]\d{1,2}$')
);

-- Stores
CREATE TABLE Stores (
    StoreID SERIAL PRIMARY KEY,
    StoreName VARCHAR(100) NOT NULL,
    Address TEXT NOT NULL,
    City VARCHAR(50) NOT NULL,
    State CHAR(2) NOT NULL,
    ZipCode VARCHAR(10) NOT NULL,
    PhoneNumber VARCHAR(20) NOT NULL
);

-- Employees
CREATE TABLE Employees (
    EmployeeID SERIAL PRIMARY KEY,
    StoreID INT NOT NULL REFERENCES Stores(StoreID),
    FirstName VARCHAR(50) NOT NULL,
    LastName VARCHAR(50) NOT NULL,
    Title VARCHAR(50) NOT NULL,
    PhoneNumber VARCHAR(25) NOT NULL,
    Email VARCHAR(100) NOT NULL,
    Salary DECIMAL(10,2) NOT NULL CHECK (Salary > 0),
    HireDate TIMESTAMPTZ,
    TerminationDate TIMESTAMPTZ
);

-- Store Managers
CREATE TABLE StoreManagers (
    StoreID INT REFERENCES Stores(StoreID) ON DELETE CASCADE,
    ManagerID INT REFERENCES Employees(EmployeeID) ON DELETE CASCADE,
    PRIMARY KEY (StoreID, ManagerID)
);

-- Customers
CREATE TABLE Customers (
    CustomerID SERIAL PRIMARY KEY,
    PhoneNumber VARCHAR(25) NOT NULL,
    Email VARCHAR(100) CHECK (Email ~* '^[A-Za-z0-9._%-]+@[A-Za-z0-9.-]+[.][A-Za-z]+$'),
    JoinedDate TIMESTAMPTZ NOT NULL DEFAULT NOW(),
    LoyaltyPoints INT NOT NULL DEFAULT 0,
    LastPurchaseDate TIMESTAMPTZ
);

-- Deliveries
CREATE TABLE Deliveries (
    DeliveryID SERIAL PRIMARY KEY,
    VendorID INT NOT NULL REFERENCES Vendors(VendorID),
    StoreID INT NOT NULL REFERENCES Stores(StoreID),
    DeliveryDate TIMESTAMPTZ NOT NULL,
    TotalCost DECIMAL(10,2) NOT NULL CHECK (TotalCost > 0),
    ReceivedBy INT REFERENCES Employees(EmployeeID),
    Status VARCHAR(20) DEFAULT 'Pending' CHECK (Status IN ('Pending', 'Delivered', 'Partial', 'Cancelled'))
);

-- Delivery Details
CREATE TABLE DeliveryDetails (
    DeliveryDetailID SERIAL PRIMARY KEY,
    DeliveryID INT NOT NULL REFERENCES Deliveries(DeliveryID) ON DELETE CASCADE,
    ProductID INT NOT NULL REFERENCES Products(ProductID),
    Quantity INT NOT NULL CHECK (Quantity > 0),
    UnitCost DECIMAL(10,2) NOT NULL CHECK (UnitCost > 0),
    ExpiryDate DATE
);

-- Promotions
CREATE TABLE Promotions (
    PromotionID SERIAL PRIMARY KEY,
    PromotionName VARCHAR(100) NOT NULL,
    Description TEXT,
    StartDate TIMESTAMPTZ NOT NULL,
    EndDate TIMESTAMPTZ NOT NULL,
    DiscountPercentage DECIMAL(5,2) CHECK (DiscountPercentage BETWEEN 0 AND 100),
    ProductID INT REFERENCES Products(ProductID),
    IsActive BOOLEAN NOT NULL DEFAULT TRUE,
    CONSTRAINT valid_dates CHECK (EndDate >= StartDate)
);

-- Sales
CREATE TABLE Sales (
    SaleID SERIAL PRIMARY KEY,
    StoreID INT NOT NULL REFERENCES Stores(StoreID),
    EmployeeID INT NOT NULL REFERENCES Employees(EmployeeID),
    CustomerID INT REFERENCES Customers(CustomerID),
    PromotionID INT REFERENCES Promotions(PromotionID),
    SaleDate TIMESTAMPTZ NOT NULL DEFAULT NOW(),
    TotalAmount DECIMAL(10,2) NOT NULL CHECK (TotalAmount >= 0),
    PaymentMethod VARCHAR(20) CHECK (PaymentMethod IN ('Cash', 'Credit', 'Debit', 'EBT', 'Mobile')),
    ChangeGiven DECIMAL(10,2) DEFAULT 0
);

-- Sale Details
CREATE TABLE SaleDetails (
    SaleDetailID SERIAL PRIMARY KEY,
    SaleID INT NOT NULL REFERENCES Sales(SaleID) ON DELETE CASCADE,
    ProductID INT NOT NULL REFERENCES Products(ProductID),
    Quantity INT NOT NULL CHECK (Quantity > 0),
    UnitPrice DECIMAL(10,2) NOT NULL CHECK (UnitPrice >= 0),
    DiscountAmount DECIMAL(10,2) DEFAULT 0,
    CONSTRAINT unique_sale_product UNIQUE (SaleID, ProductID)
) WITH (FILLFACTOR = 90);

-- Inventory
CREATE TABLE Inventory (
    InventoryID SERIAL PRIMARY KEY,
    StoreID INT NOT NULL REFERENCES Stores(StoreID),
    ProductID INT NOT NULL REFERENCES Products(ProductID),
    Quantity INT NOT NULL CHECK (Quantity >= 0),
    LowStockThreshold INT NOT NULL DEFAULT 10,
    ReorderQuantity INT NOT NULL DEFAULT 50,
    LastStockTake TIMESTAMPTZ,
    UNIQUE (StoreID, ProductID)
);

-- Employee Shift
CREATE TABLE EmployeeShift (
    ShiftID SERIAL PRIMARY KEY,
    EmployeeID INT NOT NULL REFERENCES Employees(EmployeeID),
    ShiftDate DATE NOT NULL,
    ShiftStart TIMESTAMPTZ NOT NULL,
    ShiftEnd TIMESTAMPTZ NOT NULL,
    BreakMinutes INT DEFAULT 30,
    CONSTRAINT valid_shift CHECK (ShiftEnd > ShiftStart)
);

-- Payments
CREATE TABLE Payments (
    PaymentID BIGSERIAL PRIMARY KEY,
    PaymentType VARCHAR(10) NOT NULL CHECK (PaymentType IN ('Vendor', 'Customer', 'Refund')),
    VendorID INT REFERENCES Vendors(VendorID),
    SaleID INT REFERENCES Sales(SaleID),
    DeliveryID INT REFERENCES Deliveries(DeliveryID),
    PaymentDate TIMESTAMPTZ NOT NULL DEFAULT NOW(),
    PaymentMethod VARCHAR(20) NOT NULL CHECK (PaymentMethod IN ('Cash', 'Credit', 'Debit', 'EBT', 'Mobile')),
    Amount DECIMAL(10,2) NOT NULL CHECK (Amount != 0),
    ReferenceNumber VARCHAR(50),
    CONSTRAINT valid_payment_type CHECK (
        (PaymentType = 'Vendor' AND VendorID IS NOT NULL AND SaleID IS NULL) OR
        (PaymentType = 'Customer' AND SaleID IS NOT NULL AND VendorID IS NULL) OR
        (PaymentType = 'Refund' AND SaleID IS NOT NULL)
    )
);
"""


with conn:
    with conn.cursor() as cur:
        cur.execute(create_tables_sql)
        print("Tables created successfully!")
```
